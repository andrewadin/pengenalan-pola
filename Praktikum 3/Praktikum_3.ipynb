```python
data = pd.read_csv("gdrive/My Drive/Data/tips.csv")
plt.scatter(data['day'], data['tip'])
plt.title('Scatter Plot')
plt.xlabel('Day')
plt.ylabel('Tip')

plt.show()
```

```python
plt.scatter(data['smoker'], data['tip'])
plt.title('Scatter Plot')
plt.xlabel('Smoker')
plt.ylabel('Tip')

plt.show()
```

```python
plt.scatter(data['size'], data['tip'])
plt.title('Scatter Plot')
plt.xlabel('Size')
plt.ylabel('Tip')

plt.show()
```

```python
plt.scatter(data['sex'], data['tip'])
plt.title('Scatter Plot')
plt.xlabel('Sex')
plt.ylabel('Tip')

plt.show()
```

```python
plt.plot(data['tip'])
plt.plot(data['size'])

plt.title("Scatter Plot")

plt.xlabel('Day')
plt.ylabel('Tip')

plt.show()
```

```python
plt.bar(data['day'], data['tip'])

plt.title("Bar Chart")
```

```python
plt.hist(data['total_bill'])
plt.title("Histogram")
plt.show()
```
